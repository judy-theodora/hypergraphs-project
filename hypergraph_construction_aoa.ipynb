## Construct hypergraph

```python
graph_data_aoa = pickle.load(open('./pickle/gdata_aoa.p','rb'))

word_data_9 = pickle.load(open('./pickle/word_data_aoa_9.p','rb'))
```

```python
hdata = graph_data_aoa[['word','age','logfreq','len']]
```

```python
word_data_split = []
# pair each word with a number describing the hyperedge it belongs to
# each hyperedge is a single 4-word association row from the SWOW dataset
# each unique word can appear in multiple hyperedges
for idx, word_rows in enumerate(word_data_9):
    word_data_split += [[idx,w] for w in word_rows]
```

```python
# add the word data to a pandas dataframe
word_data_frame = pd.DataFrame(word_data_split)
```

```python
# construct hypergraph from the word data
a = time()
HT = hnx.Hypergraph(word_data_frame)
print(time()-a)
```

```python
HT.save_state('./graphs/hypergraph_age_9.h')
```

## Calculate degree for every node

```python
nodes = list(HT.nodes)
```

```python
hgraph_degrees = dict()
```

```python
for node in nodes:
    hgraph_degrees[node] = HT.degree(node, s=1)
```

```python
# map degree values onto corresponding words in the dataframe
hdata['degree'] = hdata['word'].map(hgraph_degrees)
```
